## Question 2:
What potential issues or data quality concerns does the data surface? 
(These could be anomalies, missing data, inconsistent definitions, etc.)

Based on the data set provided, there are several data quality concerns 
that fall into the the data quality pillars including completeness, consistency, validity, and usability.

### Risk in nulls:
Various columns from the source tables including created date, user_id, event_type, and num_docs 
which are mandtory for downstream metrics analysis could be null which effect metrics and downstream models
including firm usage, user_engagement, and daily events summary

### No further dim tables and lack of data dictionary
It would me more managable and useful to have more dimensions that tie to event_types and their description.

### No Dim history or utilization of slow changing dimensions
We currently do not have historical tracking for users and firms dim. This means any attribute that is changes, 
we would not have any historical context. This has impacts to models as metadata changes impacting 
changes in ARR, rerention, and user title changes. SCD 6 recommended for tracking changes and current status.

### Cannot Gurantee fields are correct:
We have no confirmation that the fields in the raw data at this time are accurate.
For example:
What if Feedback score is suppose to be between 1-4 however we see 5 as a value in dataset?
What if we see arr equaling 0 in the firms table?
We need data quaility checks to ultimtely confirm if the downstream models created are accurate

### No Sessionization
We currently do not track how long a user took actions given a session, or how many sessions per product type.
This would allow more accurate power user definitions and understanding stickiness. It would also allow more anomaly 
detection in the scenario for example we see 100 document uploads and query_types within 7 seconds.

### Missing Events
We cannot gurantee from the raw we are missing events or if we hve duplicates as time granulatiry is at a date level 
and not datetime. We would require more accurate datetime and completeness check as validations

### Missing Error or Warnings in interactions
Currently we do not have an error table when user is querying, uploading documents or accessing a product.
This would help in monitoring friction, monitor system stability, and improve the product environment.

### Missing some metadata
We do not have useful user metadata such as if a user is active/unactive, email_id, geography.
This would impact analysis downstream like segementation and accurate active user tracking.


## In Summary
I would strongly recommend for any valuable analytics to have more dimensional modelling and data dictionay to align closer to business goals and definitions to accurately have lineage from the metrics models downstream. A data dictionary that can be managed by certain users/stakeholders in defining what a power user, active user, or various product and features metadata we are tracking. Lastly a solid data quality framework end to end to stand proper schema, anomaly detections and enable relibaility and accuracy of downstream analytics.
